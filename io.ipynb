```r
# Instalar pacotes necessários
if(!require("tidyverse")) install.packages("tidyverse")
if(!require("viridis")) install.packages("viridis")
if(!require("reshape2")) install.packages("reshape2")
if(!require("ggcorrplot")) install.packages("ggcorrplot")
if(!require("plotly")) install.packages("plotly")
if(!require("effectsize")) install.packages("effectsize")
if(!require("ez")) install.packages("ez")
if(!require("corrr")) install.packages("corrr")

# Carregar bibliotecas
library(tidyverse)
library(viridis)
library(reshape2)
library(ggcorrplot)
library(plotly)
library(effectsize)
library(ez)
library(corrr)

# Dataset como string
data_string <- "ID,Group,Age,Gender,Programming_Experience,Anxiety_State,Self_Efficacy,Condition_LLM,Condition_Blend,task_1,task_2,task_3,task_4,task_5,task_6,task_7,task_8,task_9,task_10,Eye_Fixation_Duration,Eye_Fixation_Count,Eye_Saccade_Amplitude,Eye_Pupil_Diameter
1,1,22,1,4,45,92,1,1,6.8,7.5,7.2,8.1,7,7.9,7.5,7.3,8.3,8,350,180,7.2,4.5
2,1,24,2,3,42,95,1,1,7.2,7.9,7.6,8.5,7.4,8.3,7.9,7.7,8.7,8.4,320,165,6.8,4.2
3,1,21,1,5,48,90,1,1,6.5,7.2,7,7.8,6.8,7.6,7.2,7,8,7.7,380,195,7.8,4.8
4,1,23,2,4,40,97,1,1,7.5,8.2,7.9,8.8,7.7,8.6,8.2,8,9,8.7,300,155,6.4,4
5,1,20,1,3,46,93,1,1,7,7.7,7.4,8.3,7.2,8.1,7.7,7.5,8.5,8.2,340,175,7,4.4
6,2,25,1,4,52,88,1,0,6.2,6.9,6.6,7.5,6.4,7.3,6.9,6.7,7.7,7.4,400,210,8.4,5.2
7,2,23,2,3,49,92,1,0,6.6,7.3,7,7.9,6.8,7.7,7.3,7.1,8.1,7.8,360,185,7.6,4.7
8,2,21,1,5,50,85,1,0,6,6.7,6.4,7.2,6.2,7,6.7,6.5,7.4,7.1,420,220,8.8,5.5
9,2,24,2,4,47,95,1,0,7,7.7,7.4,8.3,7.2,8.1,7.7,7.5,8.5,8.2,330,170,7,4.3
10,2,22,1,3,54,90,1,0,6.3,7,6.7,7.6,6.5,7.4,7,6.8,7.8,7.5,390,205,8.2,5.1
11,3,23,2,3,50,85,0,1,6.1,6.8,6.5,7.3,6.3,7.1,6.8,6.6,7.6,7.3,410,215,8.6,5.4
12,3,20,1,4,48,88,0,1,6.4,7.1,6.8,7.7,6.6,7.5,7.1,6.9,7.9,7.6,370,190,7.7,4.9
13,3,24,2,5,52,82,0,1,5.9,6.6,6.3,7,6.1,6.9,6.6,6.4,7.3,7,430,225,9,5.7
14,3,21,1,4,46,90,0,1,6.7,7.4,7.1,8,6.9,7.8,7.4,7.2,8.2,7.9,350,180,7.2,4.5
15,3,25,2,3,51,85,0,1,6.2,6.9,6.6,7.5,6.4,7.3,6.9,6.7,7.7,7.4,400,210,8.4,5.2
16,4,22,2,4,56,78,0,0,5.7,6.4,6.1,6.8,5.9,6.7,6.4,6.2,7.1,6.8,450,235,9.4,6
17,4,24,1,3,54,82,0,0,6,6.7,6.4,7.2,6.2,7,6.7,6.5,7.4,7.1,420,220,8.8,5.5
18,4,21,2,5,55,75,0,0,5.5,6.2,5.9,6.6,5.7,6.5,6.2,6,7,6.7,470,245,9.8,6.3
19,4,23,1,4,52,80,0,0,6.3,7,6.7,7.6,6.5,7.4,7,6.8,7.8,7.5,390,205,8.2,5.1
20,4,20,2,3,57,77,0,0,5.8,6.5,6.2,7,6,6.8,6.5,6.3,7.2,6.9,440,230,9.2,5.9
21,1,23,1,3,43,94,1,1,7.1,7.8,7.5,8.4,7.3,8.2,7.8,7.6,8.6,8.3,330,170,6.9,4.3
22,2,21,2,4,46,91,1,0,6.9,7.6,7.3,8.2,7.1,8,7.6,7.4,8.4,8.1,350,180,7.3,4.6
23,3,25,1,5,44,93,0,1,7.4,8.1,7.8,8.7,7.6,8.5,8.1,7.9,8.9,8.6,310,160,6.6,4.1
24,4,22,2,2,51,83,0,0,6.1,6.8,6.5,7.4,6.3,7.2,6.8,6.6,7.6,7.3,400,205,8.3,5.3
25,1,20,2,4,42,96,1,1,7.6,8.3,8,9,7.8,8.7,8.3,8.1,9.1,8.8,290,150,6.3,3.9
26,2,24,1,3,49,90,1,0,6.5,7.2,6.9,7.8,6.7,7.6,7.2,7,8,7.7,370,190,7.8,4.8
27,3,21,2,2,47,92,0,1,7.1,7.8,7.5,8.4,7.3,8.2,7.8,7.6,8.6,8.3,330,170,6.9,4.3
28,3,23,1,3,50,85,0,0,6.4,7.1,6.8,7.7,6.6,7.5,7.1,6.9,7.9,7.6,410,210,8.5,5.4
29,1,22,2,5,41,98,1,1,7.9,8.6,8.3,9.2,8.1,8.9,8.6,8.4,9.4,9.1,270,140,6,3.7
30,2,25,1,4,45,93,1,0,7.2,7.9,7.6,8.5,7.4,8.3,7.9,7.7,8.7,8.4,340,175,7.1,4.5
31,3,23,2,3,43,95,0,1,7.5,8.2,7.9,8.8,7.7,8.6,8.2,8,9,8.7,300,155,6.5,4
32,4,21,1,2,53,80,0,0,5.9,6.6,6.3,7.1,6.1,6.9,6.6,6.4,7.3,7,430,220,9.1,5.8
33,1,24,1,4,44,96,1,1,7.7,8.4,8.1,9.1,7.9,8.8,8.4,8.2,9.2,8.9,280,145,6.1,3.8
34,2,22,2,3,47,92,1,0,7,7.7,7.4,8.3,7.2,8.1,7.7,7.5,8.5,8.2,360,185,7.5,4.7
35,3,20,1,5,42,94,0,1,7.3,8,7.7,8.6,7.5,8.4,8,7.8,8.8,8.5,320,165,6.7,4.2
36,4,23,2,4,51,84,0,0,6.2,6.9,6.6,7.5,6.4,7.3,6.9,6.7,7.7,7.4,400,210,8.4,5.3
37,1,21,2,2,45,91,1,1,7.1,7.8,7.5,8.4,7.3,8.2,7.8,7.6,8.6,8.3,340,175,7,4.4
38,2,25,1,3,43,95,1,0,7.4,8.1,7.8,8.7,7.6,8.5,8.1,7.9,8.9,8.6,330,170,6.9,4.3
39,3,22,2,4,41,97,0,1,7.6,8.3,8,9,7.8,8.7,8.3,8.1,9.1,8.8,290,150,6.3,3.9
40,4,20,1,3,52,82,0,0,6,6.7,6.4,7.3,6.2,7.1,6.7,6.5,7.5,7.2,420,215,8.7,5.6"

# Ler o dataset
df <- read.csv(text = data_string)

# Normalizar variáveis neurofisiológicas para evitar escalas desiguais
df_normalized <- df %>%
  mutate_at(vars(Eye_Fixation_Duration, Eye_Fixation_Count, Eye_Saccade_Amplitude, Eye_Pupil_Diameter), scale)

# Adicionar rótulos de condições experimentais
df_normalized$LLM_Label <- factor(ifelse(df$Condition_LLM == 1, "Com LLM", "Sem LLM"))
df_normalized$Blend_Label <- factor(ifelse(df$Condition_Blend == 1, "Com Blend", "Sem Blend"))

# Definir um tema alternativo para os gráficos
theme_set(theme_minimal() + theme(plot.background = element_rect(fill = "#f0f0f0")))

# Função para gerar gráficos de correlação com ajuste de p-values
gerar_correlacao <- function(vars_selecionadas) {
  correlation_matrix <- cor(df_normalized[, vars_selecionadas])
  p_values <- cor_pmat(df_normalized[, vars_selecionadas])

  ggcorrplot(correlation_matrix, hc.order = TRUE, type = "lower", p.mat = p_values, sig.level = 0.05,
             lab = TRUE, lab_size = 3, method="circle", colors = c("#6D9EC1", "white", "#E46726"),
             title="Matriz de Correlação com P-Values Ajustados") +
    theme(plot.title = element_text(hjust = 0.5))
}

# Função para gerar gráficos de densidade para variáveis neurofisiológicas
gerar_grafico_densidade <- function(var1, var2) {
  ggplot(df_normalized, aes(x = .data[[var1]], fill = interaction(LLM_Label, Blend_Label))) +
    geom_density(alpha = 0.7) +
    scale_fill_viridis(discrete = TRUE, option = "magma") +
    labs(title = paste("Comparação de", var1, "e", var2),
         x = var1,
         y = "Densidade") +
    theme(legend.position = "bottom", legend.title = element_blank())
}

# Função para gerar boxplots comparando neurofisiologia e escalas
gerar_grafico_caixa <- function(var1, var2) {
  ggplot(df_normalized, aes(x = interaction(LLM_Label, Blend_Label), y = .data[[var1]], fill = interaction(LLM_Label, Blend_Label))) +
    geom_boxplot(alpha = 0.7) +
    scale_fill_viridis(discrete = TRUE, option = "magma") +
    labs(title = paste("Boxplot de", var1, "e", var2),
         x = "Condição Experimental",
         y = var1) +
    theme(legend.position = "bottom", legend.title = element_blank())
}

# Função para gerar heatmaps para variáveis neurofisiológicas
gerar_mapa_calor <- function(var) {
  df_heatmap <- df_normalized %>%
    select(LLM_Label, Blend_Label, all_of(var)) %>%
    group_by(LLM_Label, Blend_Label) %>%
    summarize(media = mean(.data[[var]]), .groups = "drop") %>%
    pivot_wider(names_from = Blend_Label, values_from = media)

  ggplot(df_heatmap, aes(x = LLM_Label, y = `Com Blend`, fill = media)) +
    geom_tile() +
    scale_fill_viridis(option = "magma") +
    labs(title = paste("Mapa de Calor de", var),
         x = "Uso de LLM",
         y = "Consumo de Blend") +
    theme(legend.position = "bottom", legend.title = element_blank())
}

# Geração de 30 gráficos: comparações de duas escalas ou neurofisiologia
variables <- c("Anxiety_State", "Self_Efficacy", "Eye_Fixation_Duration", "Eye_Fixation_Count", "Eye_Saccade_Amplitude", "Eye_Pupil_Diameter")

graficos_densidade <- list()
graficos_caixa <- list()
mapas_calor <- list() # Inicializar a lista mapas_calor

# Gerar gráficos de densidade e boxplots para pares de variáveis
for (i in 1:(length(variables)-1)) {
  for (j in (i+1):length(variables)) {
    graficos_densidade[[paste(variables[i], variables[j], sep = "_vs_")]] <- gerar_grafico_densidade(variables[i], variables[j])
    graficos_caixa[[paste(variables[i], variables[j], sep = "_vs_")]] <- gerar_grafico_caixa(variables[i], variables[j])
  }
}

# Gerar heatmaps para variáveis neurofisiológicas
for (var in variables) {
  mapas_calor[[var]] <- gerar_mapa_calor(var)
}

# Exibir os gráficos
print(graficos_densidade)
print(graficos_caixa)
print(mapas_calor)

# Calcular a matriz de correlação com ajuste de p-values
correlacao <- gerar_correlacao(c(variables, paste0("task_", 1:10)))
print(correlacao)

# ANOVA exemplo para a primeira tarefa - Corrigido para usar between
anova_result <- ezANOVA(
  data = df_normalized,
  dv = .(task_1),  # Variável dependente
  wid = .(ID),     # ID do participante
  between = .(LLM_Label, Blend_Label),  # Condições experimentais como fator between
  type = 3,  # Tipo III soma de quadrados
  detailed = TRUE
)
print(anova_result)

# Tamanho do efeito
eta_squared(anova_result)
```
